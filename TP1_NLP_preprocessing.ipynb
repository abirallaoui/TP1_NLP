```python
#Suppression des ponctuations

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

NewDf["text_wo_punct"] = text.apply(lambda text: remove_punctuation(text))
NewDf.head()
```

```python
#Reload the stopwords module:
from nltk.corpus import stopwords
```

```python
#Join the stopwords with commas:
", ".join(stopwords.words('english'))
```

```python
#Suppression des mots vides (stopwords)

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

NewDf["text_wo_stop"] = NewDf["text_wo_punct"].apply(lambda text: remove_stopwords(text))
NewDf.head()
```

```python
#Suppression des mots frÃ©quents

from collections import Counter
cnt = Counter()
for text in NewDf["text_wo_stop"].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

NewDf["text_wo_stopfreq"] = NewDf["text_wo_stop"].apply(lambda text: remove_freqwords(text))
NewDf.head()
```

```python
NewDf["text_wo_stopfreq"] =NewDf["text_wo_stopfreq"].str.lower()
NewDf.head()
```

```python
#Suppression des mots rares
n_rare_words = 10
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

NewDf["text_wo_stopfreqrare"] = NewDf["text_wo_stopfreq"].apply(lambda text: remove_rarewords(text))
NewDf.head()
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

NewDf["text_stemmed"] = NewDf["text_wo_stopfreqrare"].apply(lambda text: stem_words(text))
NewDf.head()
```

```python
#Lemmatization
import spacy

nlp = spacy.load("en_core_web_sm")

def lemmatize_words(text):
    doc = nlp(text)
    lemmatized_words = [token.lemma_ for token in doc]
    return " ".join(lemmatized_words)

NewDf["text_lemmatized"] = NewDf["text_stemmed"].apply(lambda text: lemmatize_words(text))
NewDf.head()
```

```python
#Suppression des Ã©mojis
#Reference : https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
NewDf["txt_wo_Emojis"] = NewDf["text_lemmatized"].apply(lambda text: remove_emoji(text))
```
